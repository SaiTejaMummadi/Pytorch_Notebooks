# Reshaping the Tensor

## Reshaping, stacking, squeezing and unsqueezing

* Reshaping - reshapes an input tensor to a defined shape
* View - Returns a view of an input tensor of certain shape but keep the same memory as the original
* Stacking - combine multiple tensors on top of each other
* Squeeze - Removes all '1' dimensions from tensors
* Unsqueeze - add a '1' dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted in certain way

```python
import torch

x = torch.arange(1.,10)
x, x.shape
```

```python
x.reshape(1,9)
```

```python
x.reshape(9,1).ndim
```

```python
x.ndim
```

```python
# Change the view
z = x.view(1,9)
z, z.shape
```

```python
# z and x share same memory,
z[:,0] = 5
z,x
```

```python
z[0,2] = 99
```

```python
x
```

```python
z
```

```python
# Stack tensors on top of each other

newx = torch.stack((x,x,x,x), dim=1)
```

```python
newx
```

```python
torch.hstack()
```

```python
newx = torch.hstack((x,x,x,x))
newx
```

```python
newx = torch.vstack((x,x,x,x))
newx
```

```python

a = torch.rand((10))
a
```

```python
a2 = torch.unsqueeze(a,1)
```

```python
a.shape, a2.shape
```

```python
torch.squeeze(a2,1)
```

```python
# torch.permute
#This is a view, therefore shares the original shape

xori = torch.rand((225,225,3))
xori.shape
```

```python
xnew = torch.permute(xori, (2,0,1))
```

```python
xnew.shape
```
